```python
import pandas as pd 
import numpy as np
df = pd.read_csv("C:\\Users\\Dell\\Documents\\python\\Githup\\Boda1515\\Amazon_project\\headphones-data\\Headphones.csv")
df
```

```python
df.columns
```

```python
# Step 1: Replace unwanted strings with NaN
df['Rate'] = df['Rate'].replace(["", "Previous page", " of related Sponsored Products"], np.nan)

# Step 2: Clean the 'Rate' column to keep only numeric values
df['Rate'] = df['Rate'].str.extract(r'(\d+\.\d+|\d+)')  # Extract numeric values

# Step 3: Convert the 'Rate' column to float
df['Rate'] = df['Rate'].astype(float)

df.Rate
```

```python
# To display all columns and rows  in the DataFrame
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
# To know how many data type this column contain.
data_type_counts = df['Price'].apply(lambda x: type(x).__name__).value_counts()

data_type_counts

# # Reset to default for columns
# pd.reset_option('display.max_columns')
# # Reset to default for rows
# pd.reset_option('display.max_rows')
```

```python
# Convert the 'Price' column to numeric, forcing errors to NaN
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')

# Optional: Fill NaN values with a default value (e.g., 0) if needed
df['Price'].fillna(0, inplace=True)

# Check the data types again
data_type_counts = df['Price'].apply(lambda x: type(x).__name__).value_counts()
data_type_counts
```

```python
# Step 1: Find the minimum price for each rate
min_price_df = df.groupby("Rate")['Price'].min().reset_index()

# Step 2: Merge with the original DataFrame to get other columns
result = pd.merge(df, min_price_df, on=['Rate', 'Price'])

# Step 3: Sort by 'Rate'
result_sorted = result.sort_values(by='Rate', ascending=False)

result_sorted["Image URL"]
# Display full content of a specific cell
result_sorted.loc[50, 'Image URL']

# Display full content of the entire 'Image URL' column
for url in df['Image URL']:
    print(url)
```
